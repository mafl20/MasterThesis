```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma
```

```python
k_values = {
    "Bearing": 0.86,
    "Fan": 11.73,
    "Gearbox": 0.77,
    "Slider": 1.24,
    "ToyCar": 2.75,
    "ToyTrain": 0.12,
    "Valve": 4.41
}

theta_values = {
    "Bearing": 4.08,
    "Fan": 0.5,
    "Gearbox": 2.39,
    "Slider": 2.14,
    "ToyCar": 0.51,
    "ToyTrain": 1.04,
    "Valve": 0.18
}
```

```python
# Generate x values for plotting
x = np.linspace(0, 30, 1000)
```

```python
plt.figure(figsize=(8, 5))
for machine, k in k_values.items():
    theta = theta_values[machine]
    pdf = gamma.pdf(x, a=k, scale=theta)
    plt.plot(x, pdf, label=f"{machine} (k={k}, θ={theta})", linewidth=2)  # Set linewidth

# Customize the plot
plt.title("Gamma Distributions for Different Machine Types", fontsize=16)
plt.xlabel("x", fontsize=12)
plt.ylabel("Probability Density", fontsize=12)
plt.legend(title="Machine Type", fontsize=10)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
for machine, k in k_values.items():
    theta = theta_values[machine]
    pdf = gamma.pdf(x, a=k, scale=theta)
    plt.plot(x, np.log1p(pdf), label=f"{machine} (k={k}, θ={theta})", linewidth=2)
plt.title("Gamma Distributions with Log-Scaled Density", fontsize=16)
plt.xlabel("x", fontsize=12)
plt.ylabel("log(Density)", fontsize=12)
plt.legend(title="Machine Type", fontsize=10)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
for machine, k in k_values.items():
    theta = theta_values[machine]
    pdf = gamma.pdf(x, a=k, scale=theta)
    plt.plot(np.log1p(x), pdf, label=f"{machine} (k={k}, θ={theta})", linewidth=2)
plt.title("Log-Transformed Gamma Distributions", fontsize=16)
plt.xlabel("log(x)", fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.legend(title="Machine Type", fontsize=10)
plt.grid(True)
plt.tight_layout()
plt.show()
```
