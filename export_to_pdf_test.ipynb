```python
from matplotlib.backends.backend_pdf import PdfPages
import matplotlib.pyplot as plt
import yaml

# Load the hyperparameters from the YAML file
with open("hyper_parameters.yaml", "r") as file:
    hyperparameters = yaml.safe_load(file)

# Create a string representation of the YAML content
hyperparameters_str = yaml.dump(hyperparameters, default_flow_style=False)

# Create a figure for the YAML content
fig4, ax4 = plt.subplots()
ax4.axis('off')  # Turn off the axis
ax4.text(0.5, 0.5, hyperparameters_str, fontsize=10, va='center', ha='center', wrap=True)
ax4.set_title("Hyperparameters")

# Create example figures
fig1, ax1 = plt.subplots()
ax1.plot([1, 2, 3], [4, 5, 6])
ax1.set_title("Loss Curve")

fig2, ax2 = plt.subplots()
ax2.plot([1, 2, 3], [6, 5, 4])
ax2.set_title("Anomaly Score")

fig3, ax3 = plt.subplots()
ax3.imshow([[1, 2], [3, 4]])
ax3.set_title("Confusion Matrix")

# Create a new figure with subplots for combining multiple figures
combined_fig, axes = plt.subplots(2, 2, figsize=(12, 8))  # 2x2 grid
axes = axes.flatten()  # Flatten the 2D array to a 1D list

# Add existing figures to the new figure's subplots
axes[0].plot([1, 2, 3], [4, 5, 6])  # Loss Curve
axes[0].set_title("Loss Curve")

axes[1].plot([1, 2, 3], [6, 5, 4])  # Anomaly Score
axes[1].set_title("Anomaly Score")

axes[2].imshow([[1, 2], [3, 4]])  # Confusion Matrix
axes[2].set_title("Confusion Matrix")

# Turn off the last subplot if not used
axes[3].axis('off')  # Turn off the unused subplot

# Save the combined figure to a PDF
with PdfPages("combined_training_results.pdf") as pdf:
    pdf.savefig(combined_fig)
    pdf.savefig(fig4)  # Save the hyperparameters figure
    plt.show()  # Show in notebook
```
